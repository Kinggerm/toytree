### Issue 39: random tips for random trees

```python
import toytree
```

```python
rtrees = [
    toytree.rtree.rtree(10, random_names=True),
    toytree.rtree.unittree(10, random_names=True),
    toytree.rtree.baltree(10, random_names=True),
    toytree.rtree.imbtree(10, random_names=True),
    toytree.rtree.bdtree(10, random_names=True),
]

mtree = toytree.mtree(rtrees)
mtree.draw(nrows=1, ncols=5, layout='d', width=900, height=200);
```
